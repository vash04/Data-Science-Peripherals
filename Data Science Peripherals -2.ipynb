```python
genre.agg({
    "Runtime":["mean","max"],
    "IMDB_Rating":"mean",
    "No_of_Votes":["sum","min"],
    "Gross":"sum",
    "Metascore":"min"
})
```

```python
genre[["Runtime","IMDB_Rating","No_of_Votes","Gross","Metascore"]].agg(["min","max","mean"])
```

```python
genre.apply(min)
```

```python
def starts_with_A(group):
    return group[group["Series_Title"].str.startswith("A")]
```

```python
genre.apply(starts_with_A)
```

```python
def genre_rank(group):
    group["genre_rank"]=group["IMDB_Rating"].rank(ascending=False)
    return group
```

```python
genre.apply(genre_rank)
```

```python
def normalized_rating(group):
    group["normalized_rating"]=(group["IMDB_Rating"]-group["IMDB_Rating"].min())/(group["IMDB_Rating"].max()-group["IMDB_Rating"].min())
    return group
```

```python
genre.apply(normalized_rating)
```
